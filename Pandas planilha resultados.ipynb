```python
#Tratando os dados
#colunas com valores repetidos ou considerados inúteis para vizualização
#resolvi fazer uma lista com o nomes das colunas a excluir para facilitar o entendimento
colunas_para_excluir = ["DT_ENCERRAMENTO",
                        "DT_EMISSAO_BU",
                        "NR_JUNTA_APURADORA",
                        "NR_TURMA_APURADORA",
                        "QT_ELEITORES_BIOMETRIA_NH",
                        "DT_ABERTURA",
                        "DS_AGREGADAS",
                        "DS_CARGO_PERGUNTA_SECAO",
                        "DT_CARGA_URNA_EFETIVADA",
                        "CD_CARGA_2_URNA_EFETIVADA",
                        "CD_CARGA_1_URNA_EFETIVADA",
                        "NR_URNA_EFETIVADA"]
resultados = resultados.drop(colunas_para_excluir, axis=1)
resultados
```

```python
#fazendo a exclusão de todas as linhas com valor "#NULO"
valor_a_excluir = '#NULO#'
linhas_excluidas = resultados["SG_PARTIDO"] == valor_a_excluir
resultados = resultados[~linhas_excluidas]
resultados
```
